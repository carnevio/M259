### Gruppierungen

Mit pandas kann man Daten Gruppieren und so Werte über bestimmte ähnliche Zeilen berechnen.

```python
mpg_cyl = mpg.groupby('cylinders')
mpg_cyl.first() # Gibt pro Gruppe den ersten Wert aus
```

```python
# Man kann über gruppen iterieren
for name, group in mpg_cyl:
  print(name)
  print(group.shape)
```

```python
#5.6
#a
nan_rows = mpg[mpg['horsepower'].isna()]
print(nan_rows)
#b
print(mpg.groupby('cylinders')['horsepower'].mean())
#c
def fill_horsepower_with_group_mean(df):
    df = df.copy()
    means = df.groupby('cylinders')['horsepower'].mean()
    for idx, row in df[df['horsepower'].isna()].iterrows():
        cyl = row['cylinders']
        df.at[idx, 'horsepower'] = means[cyl]
    return df
```
